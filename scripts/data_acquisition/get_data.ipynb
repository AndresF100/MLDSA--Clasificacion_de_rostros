# Ordenamiento de datos


Mover los archivos a la carpeta data


```python
import shutil
import os

def move_files(source_folder, destination_folder):
    # Check if both source and destination folders exist
    if not os.path.exists(source_folder):
        print(f"Source folder '{source_folder}' does not exist.")
        return
    
    if not os.path.exists(destination_folder):
        print(f"Destination folder '{destination_folder}' does not exist.")
        return

    # Get the list of files in the source folder
    files = os.listdir(source_folder)
    
    # Move each file from the source folder to the destination folder
    for file in files:
        source_file_path = os.path.join(source_folder, file)
        destination_file_path = os.path.join(destination_folder, file)
        shutil.move(source_file_path, destination_file_path)
        print(f"Moved '{file}' to '{destination_folder}'.")

# Example usage
source_folder = "data/Face Data/Face Dataset"
destination_folder = "data"

move_files(source_folder, destination_folder)
```

A continuación se organizan los datos descargados en dos carpetas, train y test.

```python
import os
import shutil
import random

dataset_path="data"

data_folder = "data"
train_folder = "train"
test_folder = "test"
split_ratio = 0.8  # 80% for training

# Create "train" and "test" folders if they don't exist
if not os.path.exists(train_folder):
    os.mkdir(train_folder)
if not os.path.exists(test_folder):
    os.mkdir(test_folder)

# Function to split files into train and test folders
def split_files(src_folder, train_dest, test_dest, split_ratio):
    files = os.listdir(src_folder)
    random.shuffle(files)
    split_index = int(len(files) * split_ratio)
    train_files = files[:split_index]
    test_files = files[split_index:]

    for file in train_files:
        src_path = os.path.join(src_folder, file)
        dest_path = os.path.join(train_dest, file)
        shutil.move(src_path, dest_path)

    for file in test_files:
        src_path = os.path.join(src_folder, file)
        dest_path = os.path.join(test_dest, file)
        shutil.move(src_path, dest_path)


```

```python
# Obtaining the list of subfolders
sub_folders = []

for root, dirs, files in os.walk(dataset_path):
    for dir in dirs:
        folder_path = os.path.join(root, dir)
        sub_folders.append(folder_path)


sub_folders[0:5]
```

```python
# Loop through subfolders in the "data" folder
for sub_f in sub_folders:

    if os.path.exists(sub_f):
        train_subfolder = os.path.join(train_folder, sub_f.split('\\')[-1])
        test_subfolder = os.path.join(test_folder, sub_f.split('\\')[-1])
        os.mkdir(train_subfolder)
        os.mkdir(test_subfolder)
        split_files(sub_f, train_subfolder, test_subfolder, split_ratio)
```
